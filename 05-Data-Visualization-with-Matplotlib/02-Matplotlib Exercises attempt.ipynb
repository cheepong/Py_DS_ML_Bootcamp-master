## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
#attempt
fig3 = plt.figure()
ax4 = fig3.add_axes([0,0,1,1])
ax5 = fig3.add_axes([0.2,0.5,.4,.4])
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
#attempt
fig3 = plt.figure()
ax4 = fig3.add_axes([0,0,1,1])
ax4.set_xlabel("x")
ax4.set_ylabel("z")
ax4.plot(x,z)

ax5 = fig3.add_axes([0.2,0.5,.4,.4])
ax5.set_xlabel("x")
ax5.set_ylabel("y")
ax5.set_xlim([20,22])
ax5.set_ylim([30,50])
ax5.set_title("zoom")
ax5.plot(x,y)

ax4 = fig3.add_axes([0,-1.2,1,1])
ax4.set_xlabel("x")
ax4.set_ylabel("z")
ax4.plot(x,z)

ax5 = fig3.add_axes([0.2,-0.7,.4,.4])
ax5.set_xlabel("x")
ax5.set_ylabel("y")
ax5.set_xlim([20,22])
ax5.set_ylim([30,50])
ax5.set_title("zoom")
ax5.plot(x,y)
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
#attempt
fig4, newaxes = plt.subplots(nrows=1,ncols=2)
```

** Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
#attempt
fig4, newaxes = plt.subplots(nrows=1,ncols=2)
newaxes[0].plot(x,y, color="blue", ls="--", lw=3)
newaxes[1].plot(x,z, color="red", ls="-", lw=3)
plt.tight_layout()
```

** See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
#attempt
fig4, newaxes = plt.subplots(nrows=1,ncols=2, figsize=(12,2))
newaxes[0].plot(x,y, color="blue", ls="-", lw=3)
newaxes[0].set_xlabel("x")
newaxes[0].set_ylabel("y")

newaxes[1].plot(x,z, color="red", ls="--", lw=3)
plt.tight_layout()
newaxes[1].set_xlabel("x")
newaxes[1].set_ylabel("z")
plt.tight_layout()
```
