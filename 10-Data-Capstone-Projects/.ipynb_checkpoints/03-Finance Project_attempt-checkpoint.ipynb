```python
# attempt

startdate = datetime.datetime(2006,1,1)
enddate = datetime.datetime(2016,1,1)

# Bank of America
BAC = data.DataReader(name = "BAC", data_source = 'yahoo', start = startdate, end = enddate)

# Citibank
C = data.DataReader(name = "C", data_source = 'yahoo', start = startdate, end = enddate)

# Goldman Sachs
GS = data.DataReader(name = "GS", data_source = 'yahoo', start = startdate, end = enddate)

# JPMorgan Chase
JPM = data.DataReader(name = "JPM", data_source = 'yahoo', start = startdate, end = enddate)

# Morgan Stanley
MS = data.DataReader(name = "MS", data_source = 'yahoo', start = startdate, end = enddate)

# Wells Fargo
WFC = data.DataReader(name = "WFC", data_source = 'yahoo', start = startdate, end = enddate)
```

```python
# attempt
WFC.head()
```

```python
# attempt (cleaning up)
data_list = [BAC, C, GS, JPM, MS, WFC]

for data in data_list:
    data.drop(columns = "Adj Close", inplace=True)
```

** Create a list of the ticker symbols (as strings) in alphabetical order. Call this list: tickers**

```python
# attempt
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

** Use pd.concat to concatenate the bank dataframes together to a single data frame called bank_stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis you concatenate on.**

```python
# attempt
bank_stocks = pd.concat(objs = [BAC, C, GS, JPM, MS, WFC], axis = 1, keys = tickers)
```

** What is the max Close price for each bank's stock throughout the time period?**

```python
# attempt
round(bank_stocks.xs(key = 'Close', axis = 1, level = 1).max(), 2)
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
# attempt
returns = bank_stocks.xs(key = 'Close', axis = 1, level = 1).pct_change()
returns.columns.names = ['']
returns.columns = ['BAC Return', 'C Return', 'GD Return', 'JPM Return', 'MS Return', 'WFC Return']
returns.head()
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
# attempt

fig = plt.figure(figsize=[12,4])

sns.lineplot(data=bank_stocks.loc['20080101':'20081231'].xs(key = ('BAC','Close'), axis = 1), label = 'BAC CLOSE', color = 'green')
sns.lineplot(data=bank_stocks.loc['20080101':'20081231'].xs(key = ('BAC','Close'), axis = 1).rolling(window = 30).mean(), label = '30 Day Avg', color = 'blue')

plt.legend()

plt.xlim('2008-01', '2009-01')
plt.ylim(0,50)
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
# attempt

sns.heatmap(data= bank_stocks.xs(key = ('Close'), axis = 1, level = 1).corr(), annot=True)
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
# attempt
sns.clustermap(data= bank_stocks.xs(key = ('Close'), axis = 1, level = 1).corr(), annot=True)
```
